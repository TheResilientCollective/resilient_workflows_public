```python
import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.geometry import Point
from datetime import datetime, date
import os
import sys
import logging  
#from keplergl import KeplerGl
logging.basicConfig(stream=sys.stdout, level=logging.INFO, format="%(asctime)s %(message)s")
logger=logging.getLogger(__name__)
logger.setLevel(level=logging.DEBUG)

base_dir = os.getcwd()


output_dir = os.path.join(base_dir, "../../../../data/cdc/output/")
```

```python
cdc_url ="https://data.cdc.gov/api/views/x9gk-5huc/rows.csv?fourfour=x9gk-5huc&cacheBust=17140616481&accessType=DOWNLOAD"

#Reporting Area,Current MMWR Year,MMWR WEEK,Label,Current week,"Current week, flag",Previous 52 week Max,"Previous 52 weeks Max, flag",Cumulative YTD Current MMWR Year,"Cumulative YTD Current MMWR Year, flag",Cumulative YTD Previous MMWR Year,"Cumulative YTD Previous MMWR Year, flag",LOCATION1,LOCATION2,sort_order,geocode

ndss_df = pd.read_csv(
    cdc_url, delimiter=",", 
    )

ndss_df.head()
```

```python
ndss_df.dropna(inplace=True,subset=["geocode", "Current week","Current MMWR Year","MMWR WEEK" ])
gs=gpd.GeoSeries.from_wkt(ndss_df['geocode'])  
ndss_gdf = gpd.GeoDataFrame(ndss_df, 
    geometry = gpd.GeoSeries.from_wkt(ndss_df['geocode']) ,
    crs = 'EPSG:4326')
ndss_gdf["lat"] = ndss_gdf.geometry.y
ndss_gdf["lon"] = ndss_gdf.geometry.x
ndss_gdf.info()
ndss_gdf.head()
```

```python

    
ndss_gdf['date'] = ndss_gdf.apply(lambda row: date.fromisocalendar(row['Current MMWR Year'],row['MMWR WEEK'], 1), axis=1 )
ndss_gdf['date'] = pd.to_datetime(ndss_gdf['date'])
ndss_gdf.head()
```

```python
ndss_gdf.to_csv(output_dir + "ndss_gdf.csv", index=False)
ndss_gdf.to_parquet(output_dir + "ndss_gdf.parquet", index=False)
```

```python
pathogens = ndss_gdf.groupby(by=["Label"], as_index=False).agg(count=("Current week", "sum"))
pathogens
```

```python
def generatePathogenPoint(pathogen,dataframe):
    logger.info(f"Generating pathogen {pathogen}")
    filter_df = dataframe[dataframe["Label"] == pathogen]
    if len(filter_df) == 0:
        return 
    filter_df.to_file(os.path.join(output_dir,f"ndss_{pathogen}_zip.geojson" ) , driver='GeoJSON')
    filter_df.to_csv(os.path.join(output_dir,f"ndss_{pathogen}_zip.csv" ))
    filter_df.head(20)
    
pathogen = pathogens["Label"].unique().tolist()
for g in pathogen:
    g = g.replace(" ","_").replace("-","_").replace(",","_").replace(".","_").replace("/","_")
    generatePathogenPoint(g,ndss_gdf)
```
