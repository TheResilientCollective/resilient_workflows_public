```python
import pandas as pd
import datetime 
import geopandas as gpd
```

```python
today = datetime.date.today()
```

```python
import requests

url = "https://gis-public.sandiegocounty.gov/arcgis/rest/services/Hosted/SDAPCD_Complaints/FeatureServer/0/query?where=1%3D1&objectIds=&time=&geometry=&geometryType=esriGeometryEnvelope&inSR=&spatialRel=esriSpatialRelIntersects&distance=&units=esriSRUnit_Foot&relationParam=&outFields=nature_of_complaint%2C+date_received%2C+record_number%2C+record_status%2Cinvestigation_outcome%2C+response_duration__hours_%2C+x_coordinate%2C+y_coordinate&returnGeometry=true&maxAllowableOffset=&geometryPrecision=&outSR=&havingClause=&gdbVersion=&historicMoment=&returnDistinctValues=false&returnIdsOnly=false&returnCountOnly=false&returnExtentOnly=false&orderByFields=&groupByFieldsForStatistics=&outStatistics=&returnZ=false&returnM=false&multipatchOption=xyFootprint&resultOffset=&resultRecordCount=&returnTrueCurves=false&returnCentroid=false&timeReferenceUnknownClient=false&sqlFormat=none&resultType=&datumTransformation=&lodType=geohash&lod=&lodSR=&f=geojson"

df = gpd.read_file(url)

df['date']=pd.to_datetime( df['date_received'],unit='ms' )
df = df.set_index(pd.DatetimeIndex(df['date']))
start = datetime.datetime.now() -  datetime.timedelta(days=90)
print(start)
df = df[ df['date']>=start]
df.dropna(how='any', subset=['x_coordinate', 'y_coordinate','geometry'], inplace=True)
df= df[(df['x_coordinate']!=0) | (df['y_coordinate']!=0)]
df
```

```python

date_range = pd.date_range(end=today, periods=90, freq='D')

date_range
```

```python
dates = pd.Series(date_range,name='datetime')
dates = pd.to_datetime(dates,unit='ms')
dates_df = pd.DataFrame(dates)
dates_df['nature_of_complaint']='Odor'
dates_df
```

```python
joined= pd.concat([df, dates_df])
joined
```

```python
complaint_groupby_date=joined[joined['nature_of_complaint']=='Odor'].groupby(by=['datetime', 'nature_of_complaint']).resample('D').agg(count=('datetime','count'))
complaint_groupby_date
```

```python
counts_df= df[df['nature_of_complaint']=='Odor'].groupby(by=[ 'nature_of_complaint']).resample('D').agg(count=('nature_of_complaint','count'))
counts_df.reset_index()
```

```python
sra_gdf = gpd.read_file('../../../../data/geographic/sra.geojson')
sra_gdf
```

```python
df2 = df.drop(columns=['date'])

complaints_sra_gdf=sra_gdf.sjoin(df2,how='left',predicate='contains',lsuffix='_left',rsuffix='_right')
complaints_sra_gdf
```

```python
complaints_sra_gdf['date']=pd.to_datetime( complaints_sra_gdf['date_received'],unit='ms' )
complaints_sra_gdf = complaints_sra_gdf.set_index(pd.DatetimeIndex(complaints_sra_gdf['date']))

counts_sra_df= complaints_sra_gdf[complaints_sra_gdf['nature_of_complaint']=='Odor'].groupby(by=[ 'name','nature_of_complaint','geometry','x_coordinate','y_coordinate','date_received','sra','globalid']).resample('D').agg(count=('nature_of_complaint','count'))

counts_sra_df=counts_sra_df.reset_index()
counts_sra_df['date_received']=pd.to_datetime( counts_sra_df['date_received'],unit='ms')
counts_sra_df['date_received']=counts_sra_df['date_received'].dt.tz_localize("US/Pacific") 
#counts_sra_df['date']=pd.to_datetime( counts_sra_df['date_received'],unit='ms' )
#counts_sra_df.set_index(pd.DatetimeIndex(df['date']))
```
